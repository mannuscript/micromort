```python
import json
import traceback, sys
import os
import pandas as pd

from micromort.models.polarity.stanford_deeply_moving import StanfordDeeplymoving
from micromort.data_stores.mongodb import getConnection

from multiprocessing.dummy import Pool as ThreadPool 
```

```python
mongo_db_name = "micromort"
mongo_collection_name = "newstweets_categorized_news"
mongoClient = getConnection(mongo_db_name, mongo_collection_name)
```

```python
def storeInMongo(collection,  item):

    url = item["url"]
    collection.update({ 
        "url"  : url
        } 
    ,  dict(item),  True)
```

```python
% export LC_CTYPE=en_US.UTF-8
```

```python
polarity_classifier = StanfordDeeplymoving()
```

```python
def myFun(tweet):
    try:
        d = tweet["text"]
        polarity = polarity_classifier.get_sentiment_string(d.encode('ascii',errors='ignore')[:500])
        if polarity == "":
            polarity = "Neutral"

        tweet["polarity"] = polarity
        storeInMongo(mongoClient, tweet)
    except Exception as e:
        print e
    
    
pool = ThreadPool(6) 
results = pool.imap(myFun, mongoClient.find())
```
